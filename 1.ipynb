```python
import pandas as pd
import numpy as np

# Q1: From the winequality.csv extract the first 100 rows and the 12 columns
url = "https://raw.githubusercontent.com/benny472/BehnamSafayi/main/winequality.csv"
W = pd.read_csv(url,header=0, sep=";")
W1 = W.iloc[:100,:12]
W1
```

```python
# Q2: Transform in the numpy array A
A = np.array(W1)

# Q3: Calculate the dot product Y = A'A
AT = A.transpose()
Y = np.dot(AT,A)

# Q4: Print the result Y
print("Y = A'A = ", "\n", Y)
```

```python
# Q5: Reshape A as an array with dimension (?,25,12)
RA = np.reshape(A,(-1,25,12))
print(RA.shape, "\n")
print(RA)
```

```python
# Q6: Create a function able to make the dot product of two matrices and return the sum of all elements of the result
def funw(x,y):

    return np.sum(np.dot(x,y))

# example
S = funw(AT,A)
print("Sum of all elements = ", S)
```
